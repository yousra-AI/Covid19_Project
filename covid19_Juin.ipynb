# covid_19 dataset extraction and cleaning

```python
import pandas as pd
import requests
```

## extract the data from wikipedia

```python
#extract the data
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic'
req = requests.get(url)
Data_list = pd.read_html(req.text)
target_df = Data_list[4]
```

```python
target_df.head()
```

## Clean the data

```python

target_df.columns = ['col0', 'country_Name', 'Total_cases', 'Total_Deathes', 'Total_Recoveries', 'col5']
```

```python
target_df.head()
```

```python
target_df = target_df[['country_Name', 'Total_cases', 'Total_Deathes', 'Total_Recoveries']]
```

```python
target_df.head()
```

```python
#drop the last 2 rows
last_idx = target_df.index[-1]
target_df = target_df.drop([last_idx, last_idx-1])
```

```python
target_df.head(-1)
```

```python
# replace data
target_df['country_Name'] = target_df['country_Name'].str.replace('\[.*\]','')

target_df['Total_Recoveries'] = target_df['Total_Recoveries'].str.replace('No data','0')
```

```python
target_df.head()
```

```python
# change data to numeric
target_df['Total_cases'] = pd.to_numeric(target_df['Total_cases'])
target_df['Total_Deathes'] = pd.to_numeric(target_df['Total_Deathes'])
target_df['Total_Recoveries'] = pd.to_numeric(target_df['Total_Recoveries'])
```
